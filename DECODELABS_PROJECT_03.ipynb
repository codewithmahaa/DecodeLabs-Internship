```python



import re
from urllib.parse import urlparse


class PhishingAnalyzer:
    def __init__(self):

        self.suspicious_keywords = [
            "urgent", "immediately", "verify", "login", "password",
            "bank", "account suspended", "click here", "limited time",
            "security alert", "confirm identity", "update details"
        ]

    def extract_urls(self, text):
        """Extract URLs from text"""
        url_pattern = r'https?://[^\s]+'
        return re.findall(url_pattern, text)

    def check_suspicious_urls(self, urls):
        """Check for suspicious URL patterns"""
        suspicious = []
        for url in urls:
            parsed = urlparse(url)
            domain = parsed.netloc

            if any(x in url for x in ["@", "-", "bit.ly", "tinyurl"]):
                suspicious.append(url)
            elif len(domain.split(".")) > 3:
                suspicious.append(url)

        return suspicious

    def find_keywords(self, text):
        """Find phishing-related keywords"""
        found = []
        text_lower = text.lower()

        for keyword in self.suspicious_keywords:
            if keyword in text_lower:
                found.append(keyword)

        return found

    def analyze_message(self, message):
        """Main analysis function"""
        result = {
            "suspicious_links": [],
            "keywords_found": [],
            "red_flags": [],
            "explanation": ""
        }

        urls = self.extract_urls(message)
        suspicious_urls = self.check_suspicious_urls(urls)
        keywords = self.find_keywords(message)

        result["suspicious_links"] = suspicious_urls
        result["keywords_found"] = keywords

        if suspicious_urls:
            result["red_flags"].append("Contains suspicious or shortened URLs")

        if keywords:
            result["red_flags"].append("Uses urgent or sensitive keywords")

        if "@" in message:
            result["red_flags"].append("Possible fake email formatting")

        if any(word in message.lower() for word in ["click", "login", "verify"]):
            result["red_flags"].append("Attempts to make user take immediate action")

        if result["red_flags"]:
            result["explanation"] = (
                "This message appears unsafe because it contains phishing indicators such as "
                f"{', '.join(result['red_flags'])}. These tactics are commonly used to trick users "
                "into revealing sensitive information."
            )
        else:
            result["explanation"] = "No strong phishing indicators found, but remain cautious."

        return result


if __name__ == "__main__":
    analyzer = PhishingAnalyzer()

    sample_message = """
    Dear User,
    Your account has been suspended. Please click here immediately:
    http://bit.ly/fake-login to verify your account.
    """

    analysis = analyzer.analyze_message(sample_message)

    print("=== PHISHING ANALYSIS REPORT ===")
    print("Suspicious Links:", analysis["suspicious_links"])
    print("Keywords Found:", analysis["keywords_found"])
    print("Red Flags:", analysis["red_flags"])
    print("Explanation:", analysis["explanation"])
```
