```python
"""
Password Strength Checker
A tool to evaluate password security based on multiple criteria
"""

import re  # For special character detection

def check_common_passwords(password):

    common_passwords = [
        'password', '123456', '12345678', '123456789', 'qwerty',
        'abc123', 'password123', 'admin', 'letmein', 'welcome',
        'monkey', 'dragon', 'master', 'football', 'baseball'
    ]

    return password.lower() in common_passwords

def check_password_strength(password):


    # Initialize criteria checks
    length_valid = len(password) >= 8
    has_number = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_special = any(not char.isalnum() for char in password)

    criteria_met = sum([length_valid, has_number, has_upper, has_special])

    if check_common_passwords(password):
        return "WEAK", " This is a commonly used password! Very easy to guess."

    # Strength classification logic
    if length_valid and has_number and has_upper and has_special and has_lower:
        return "STRONG", " Excellent! Your password is very secure."
    elif criteria_met >= 3 and length_valid:
        return "MEDIUM", " Good, but could be stronger. Add missing elements."
    else:
        return "WEAK", " Too weak! Please improve your password."

def provide_feedback(password, strength, message):
    """
    Show detailed feedback about what's missing for weak/medium passwords
    """
    print(f"\n{'='*50}")
    print(f"PASSWORD STRENGTH: {strength}")
    print(f"{'='*50}")
    print(message)


    if strength != "STRONG":
        print("\n📋 IMPROVEMENT TIPS:")

        if len(password) < 8:
            print("  • Make it at least 8 characters long")
        if not any(char.isdigit() for char in password):
            print("  • Add at least one number (0-9)")
        if not any(char.isupper() for char in password):
            print("  • Add at least one uppercase letter (A-Z)")
        if not any(char.islower() for char in password):
            print("  • Add at least one lowercase letter (a-z)")
        if not any(not char.isalnum() for char in password):
            print("  • Add at least one special symbol (!@#$% etc.)")

    print("\ WHAT YOU DID WELL:")
    if len(password) >= 8:
        print(f"  • Length: {len(password)} characters (✓ meets requirement)")
    if any(char.isdigit() for char in password):
        print("  • Contains numbers (✓)")
    if any(char.isupper() for char in password):
        print("  • Contains uppercase letters (✓)")
    if any(char.islower() for char in password):
        print("  • Contains lowercase letters (✓)")
    if any(not char.isalnum() for char in password):
        print("  • Contains special characters (✓)")

def main():
    """
    Main program loop with input validation
    """
    print("\n" + "="*50)
    print(" PASSWORD STRENGTH CHECKER ")
    print("="*50)
    print("\nPassword requirements:")
    print("  • Minimum 8 characters")
    print("  • At least 1 number")
    print("  • At least 1 uppercase letter")
    print("  • At least 1 special symbol (!@#$% etc.)")

    while True:
        # Get password input
        password = input("\nEnter your password (or 'quit' to exit): ")

        # Exit option
        if password.lower() == 'quit':
            print("\n Goodbye! Stay secure!")
            break

        # Skip empty passwords
        if not password:
            print(" Password cannot be empty! Please try again.")
            continue

        strength, message = check_password_strength(password)

        provide_feedback(password, strength, message)

        print("\n" + "-"*50)
        another = input("Would you like to check another password? (yes/no): ").lower()
        if another != 'yes' and another != 'y':
            print("\n Goodbye! Stay secure!")
            break

if __name__ == "__main__":
    main()
```
